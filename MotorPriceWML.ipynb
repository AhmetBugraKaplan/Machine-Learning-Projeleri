```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("motor.csv")
df
```

```python
df = df.drop(columns=["name","ex_showroom_price"])
df.head()
```

```python
df.columns=["satisfiyat","yil","saticitipi","sahip","km"]
df.head(3)
```

```python
df.info()
```

```python
df = pd.get_dummies(df,columns=["saticitipi","sahip"],drop_first=False)
```

```python
df.head(3)
```

```python
df.info()
```

```python
y = df[["satisfiyat"]]
x = df.drop("satisfiyat",axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.70,random_state=79)
```

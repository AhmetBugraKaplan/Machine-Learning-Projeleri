# `Muhtemelen bu kısımda dummies yapınca yüzlerce yeni sütün oluşacak ve verimsiz olacak fakat bu durumda ne olduğunu merak ettiğim için öncelikle böyle deneyeceğim`

```python
dfWDummies = pd.get_dummies(df,columns=["Company","Product","TypeName","ScreenResolution","Cpu","Gpu","OpSys","StorageType"],drop_first=True)
```

```python
dfWDummies.head(5)
```

```python
y = dfWDummies["Price_euros"]
x = dfWDummies.drop(columns=["Price_euros"],axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=30,train_size=0.70)
```

```python
lm = LinearRegression()
model = lm.fit(x_train,y_train)
model.score(x_test,y_test)
```

# ***String veriyi olduğu için dummie yapmak zorundasın bunu hatırlamış olduk. Şimdi  RideRegrasyon ile deneyeceğim***

```python
alfalar = [0.1, 0.3, 0.5, 0.8, 1, 10, 20, 50, 100, 200]

max_skor = 0
besta = 0

for a in alfalar:
    r = Ridge(alpha=a)
    model = r.fit(x_train, y_train)
    skor = model.score(x_test, y_test)
    print(f"Alpha: {a}  →  Skor: {skor}")

    if skor > max_skor:
        max_skor = skor
        besta = a

print(f"\n En iyi skor {max_skor:.4f} alpha={besta} ile bulundu")
```

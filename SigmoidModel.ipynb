```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df  = pd.read_csv("personality.csv")
df.head(3)
```

```python
df.info()
```

```python
df["Stage_fear"].unique()
df['Drained_after_socializing'].unique()
df['Personality'].unique()
```

```python
df["Stage_fear"] = df['Stage_fear'].map({"No":0,"Yes":1})
df.head(3)
```

```python
df["Drained_after_socializing"] = df["Drained_after_socializing"].map({"No":0,"Yes":1})
df["Personality"] = df["Personality"].map({"Extrovert":0,"Introvert":1})
```

```python
df.head(3)
```

```python
df.dtypes
```

***dtypes ile verilerimizin tiplerini kontrol ediyoruz. Derin renme modelimizi oluturmak iin normalde hepsinin float olmas lazm fakat Stage_fear int bu sebeple onu floata eviricez***

```python
df = df.astype("float32")
```

```python
df.dtypes
```

```python
df.head(3)
```

```python
y = df["Personality"]
x = df.drop("Personality",axis=1)
```

```python
model = Sequential()

#input layer
model.add(Dense(32,input_dim=7,activation="relu"))

#hidden layer
model.add(Dense(32,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dense(32,activation="relu"))

#outpu layer
model.add(Dense(1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])

model.fit(x=x,y=y,epochs=8,batch_size=32,validation_split=0.2)
```
